### Auxiliar methods

```python
def to_unicode(s):
    result = s
    if type(s) != unicode:
        result = s.decode('utf-8')
    return result

def equalize_dict(keys, data_dict):
    for key in keys:
        if key not in data_dict:
            data_dict[key] = 0
    return data_dict

def member(elem, seq):
    return elem[0] in map(lambda x: x[0], seq)

def intersection(a, b):
    set_a = set(map(lambda x: x[0], a))
    set_b = set(map(lambda x: x[0], b))
    return set_a & set_b
```

### Functions

```python
def node_dict_by_type(nbunch):
    typ = a['type']
    
    nodes = {}
    if type(nbunch) == nx.classes.multidigraph.DiGraph:
        nbunch = nbunch.nodes_iter(data=True)
    for node, attrs in nbunch:
        node_type = attrs[typ]
        nodes.setdefault(node_type, [])
        nodes[node_type].append((node, attrs))
    return nodes

def node_dict_by_attr(nbunch, attr):
    nodes = {}
    if type(nbunch) == nx.classes.multidigraph.DiGraph:
        nbunch = nbunch.nodes_iter(data=True)
    for node, attrs in nbunch:
        if attr in attrs:
            value = attrs[attr]
            nodes.setdefault(value, [])
            nodes[value].append((node, attrs))
    return nodes

def freq_dict(nbunch):
    freqs = {}
    for key in nbunch.iterkeys():
        freqs[key] = len(nbunch[key])
    return freqs

def node_neighbor_list(node, graph, attr=None, value=None, label=None, typ=None):
    if attr and value:
        value = to_unicode(value)
    neighbors = []
    relations = graph[node]
    if label:
        relations = filter_relations_by_label(relations, label)
    if typ:
        relations = filter_relations_by_type(graph, relations, typ)
    neighbors_attrs = [(n, graph.node[n]) for n in relations]
    for neighbor_attrs in neighbors_attrs:
        attrs = neighbor_attrs[1]
        if attr and value:
            if attr in attrs:
                val = attrs[attr]
                val = to_unicode(val)
                if val == value:
                    neighbors.append(neighbor_attrs)
        else:
            neighbors.append(neighbor_attrs)
    return neighbors

def filter_node_dict_by_value(nbunch, attr, value):
    nodes = []
    if type(nbunch) == nx.classes.multidigraph.DiGraph:
        nbunch = nbunch.nodes_iter(data=True)
    for node, attrs in nbunch:
        if attr in attrs:
            val = attrs[attr]
            if val == value:
                nodes.append((node, attrs))
    return nodes

def filter_relations_by_label(relations, label_value):
    label_name = a['label']
    
    relations_aux = {}
    for node, attrs in relations.iteritems():
        if label_name in attrs and attrs[label_name] == label_value:
            relations_aux[node] = attrs
    return relations_aux

def filter_relations_by_type(graph, relations, type_value):
    type_name = a['type']
    
    relations_aux = {}
    for node, attrs in relations.iteritems():
        node_attrs = graph.node[node]
        if type_name in node_attrs and node_attrs[type_name] == type_value:
            relations_aux[node] = attrs
    return relations_aux
```

### Number in title associated with Spaniards

```python
ara_re = r'(?P<arabic>\d+)'
rom_re = r'(?P<roman>X?(I?(V|X)|V?(III|II|I))\.)'
ara_comp = re.compile(ara_re)
rom_comp = re.compile(rom_re)

def romanToInt(numeral):
    rlist = romanList = [(10, "X"),(9, "IX"),(5, "V"),(4, "IV"),(1, "I")]
    index = 0
    intResult = 0
    for integer, romanNumeral in rlist:
        while numeral[index : index + len(romanNumeral)] == romanNumeral:
            intResult += integer
            index += len(romanNumeral)
    return intResult

def get_number(title):
    ara_result = ara_comp.search(title)
    rom_result = rom_comp.search(title)
    if ara_result:
        number = int(ara_result.group('arabic'))
    elif rom_result:
        number = romanToInt(rom_result.group('roman'))
    else:
        number = 'Painting with no number'
    return number
```

```python
numbers = {}
paintings = node_dict_by_type(castas)[a['painting']]
for painting in paintings:
    spanish_neighbors = node_neighbor_list(
        painting[0],
        castas,
        attr=a['casta'],
        value=a['espanol'],
        label=a['depicts'],
        typ=a['character'],
    )
    if spanish_neighbors:
        title = painting[1][a['painting_title']]
        number = get_number(title)
        numbers[number] = numbers.get(number, 0) + 1
```

### Heatmap violence

```python
actions = node_dict_by_type(castas)[a['action']]
```

```python
violent_actions = [a['fighting'], a['non-fighting'], a['crying']]
```

```python
def is_violence(action):
    return action[1][a['label']] in violent_actions
```

```python
def characters_receiving(action):
    character_neighbors = node_neighbor_list(
        action[0],
        castas,
        label=a['receiving'],
        typ=a['character'],
    )
    return character_neighbors
```

```python
def characters_doing(action):
    character_neighbors = []
    for character in characters:
        action_neighbors = actions_doing(character)
        if member(action, action_neighbors):
            character_neighbors.append(character)
    return character_neighbors
```

```python
def actions_receiving(character):
    action_neighbors = []
    for action in actions:
        character_neighbors = characters_receiving(action)
        if member(character, character_neighbors):
            action_neighbors.append(action)
    return action_neighbors
```

```python
def actions_doing(character):
    action_neighbors = node_neighbor_list(
        character[0],
        castas,
        label=a['doing'],
        typ=a['action'],
    )
    return action_neighbors
```

```python
def violent_actions_receiving(character):
    action_neighbors = actions_receiving(character)
    violent_actions = filter(is_violence, action_neighbors)
    return violent_actions
```

```python
def violent_actions_doing(character):
    action_neighbors = actions_doing(character)
    violent_actions = filter(is_violence, action_neighbors)
    return violent_actions
```

```python
def painting_character(character):
    result = None
    for painting in paintings:
        character_neighbors = node_neighbor_list(
            painting[0],
            castas,
            label=a['depicts'],
            typ=a['character'],
        )
        if member(character, character_neighbors):
            result = painting
            break
    return result
```

```python
def same_character(character1, character2):
    return character1[0] == character2[0]
```

```python
def same_painting(character1, character2):
    painting1 = painting_character(character1)
    painting2 = painting_character(character2)
    return painting1 and painting2 and painting1[0] == painting2[0]
```

```python
def is_receiving_violence_from(character1, character2):
    result = False
    if not same_character(character1, character2) and same_painting(character1, character2):
        violent_actions1 = violent_actions_receiving(character1)
        violent_actions2 = violent_actions_doing(character2)
        common_violent_actions = intersection(violent_actions1, violent_actions2)
        result = len(common_violent_actions) > 0
    return result
```

```python
def is_doing_violence_to(character1, character2):
    result = False
    if not same_character(character1, character2) and same_painting(character1, character2):
        violent_actions1 = violent_actions_doing(character1)
        violent_actions2 = violent_actions_receiving(character2)
        common_violent_actions = intersection(violent_actions1, violent_actions2)
        result = len(common_violent_actions) > 0
    return result
```

```python
castas_doing_and_receiving_violence_by_casta = {}
for character1 in violent_characters:
    casta1 = character1[1][a['casta']]
    for character2 in violent_characters:
        casta2 = character2[1][a['casta']]
        cond1 = is_doing_violence_to(character1, character2)
        cond2 = is_receiving_violence_from(character2, character1)
        if cond1 and cond2:
            castas_doing_and_receiving_violence_by_casta.setdefault(casta2, {})
            castas_doing_and_receiving_violence_by_casta[casta2].setdefault(casta1, 0)
            castas_doing_and_receiving_violence_by_casta[casta2][casta1] += 1
```

```python
all_violent_castas = set()
for casta1 in castas_doing_and_receiving_violence_by_casta:
    all_violent_castas.add(casta1)
    castas_doing_violence = castas_doing_and_receiving_violence_by_casta[casta1]
    for casta2 in castas_doing_violence:
        all_violent_castas.add(casta2)
```

```python
castas_doing_and_receiving_violence_by_casta = equalize_dict(all_violent_castas, castas_doing_and_receiving_violence_by_casta)
for casta in castas_doing_and_receiving_violence_by_casta:
    castas_doing_violence = castas_doing_and_receiving_violence_by_casta[casta]
    if castas_doing_violence == 0:
        castas_doing_and_receiving_violence_by_casta[casta] = {}
for casta in castas_doing_and_receiving_violence_by_casta:
    castas_doing_violence = castas_doing_and_receiving_violence_by_casta[casta]
    castas_doing_and_receiving_violence_by_casta[casta] = equalize_dict(all_violent_castas, castas_doing_violence)
```

```python
def characters_with_object(characters):
    characters_by_object = {}
    for character in characters:
        object_neighbors = objects_associated(character)
        for obj in object_neighbors:
            label = obj[1][a['label']]
            characters_by_object.setdefault(label, [])
            characters_by_object[label].append(character)
    return characters_by_object
```

```python
def freq_characters_with_objects(characters):
    characters_by_objects = characters_with_object(characters)
    freq_characters_by_objects = freq_dict(characters_by_objects)
    freq_characters_by_objects = equalize_dict(all_objects, freq_characters_by_objects)
    return freq_characters_by_objects
```
