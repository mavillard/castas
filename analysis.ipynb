### Definitions

```python
# Attribute dictionary
a = {
    # schema
    'type': u'[Schema] Type', # node type
    'label': u'label', # relationship name
    # node types
    'action': u'Action',
    'character': u'Character',
    'painting': u'Painting',
    'setting': u'Setting',
    #relation labels
    'depict': u'depict',
    'depicts': u'depicts',
    # node attributes
    'casta': u'(Character) Casta',
    'child': u'(Character) Progeny',
    'inscription': u'(Painting) Incription',
    'male': u'(Character) Male',
    'female': u'(Character) Female',
    'parent': u'(Character) Parent',
    # relation attributes
    'doing': u'doing',
    'receiving': u'received by',
    # values
    'espanol': u'Espa\xf1ol ',
    'true': u'True',
}
```

### Auxiliar methods

```python
def filter_relations_by_label(relations, label_value):
    label_name = a['label']
    
    relations_aux = {}
    for node, attrs in relations.iteritems():
        if label_name in attrs and attrs[label_name] == label_value:
            relations_aux[node] = attrs
    return relations_aux

def filter_relations_by_type(graph, relations, type_value):
    type_name = a['type']
    
    relations_aux = {}
    for node, attrs in relations.iteritems():
        node_attrs = graph.node[node]
        if type_name in node_attrs and node_attrs[type_name] == type_value:
            relations_aux[node] = attrs
    return relations_aux

def to_unicode(s):
    result = s
    if type(s) != unicode:
        result = s.decode('utf-8')
    return result

def equalize_dict(keys, data_dict):
    for key in keys:
        if key not in data_dict:
            data_dict[key] = 0
    return data_dict
```

### API

```python
def node_dict_by_type(nbunch):
    typ = a['type']
    
    nodes = {}
    if type(nbunch) == nx.classes.multidigraph.DiGraph:
        nbunch = nbunch.nodes_iter(data=True)
    for node, attrs in nbunch:
        node_type = attrs[typ]
        nodes.setdefault(node_type, [])
        nodes[node_type].append((node, attrs))
    return nodes

def node_dict_by_attr(nbunch, attr):
    nodes = {}
    if type(nbunch) == nx.classes.multidigraph.DiGraph:
        nbunch = nbunch.nodes_iter(data=True)
    for node, attrs in nbunch:
        if attr in attrs:
            value = attrs[attr]
            nodes.setdefault(value, [])
            nodes[value].append((node, attrs))
    return nodes

def freq_dict(nbunch):
    freqs = {}
    for key in nbunch.iterkeys():
        freqs[key] = len(nbunch[key])
    return freqs

def node_neighbor_list(node, graph, attr=None, value=None, label=None, typ=None):
    if attr and value:
        value = to_unicode(value)
    neighbors = []
    relations = graph[node]
    if label:
        relations = filter_relations_by_label(relations, label)
    if typ:
        relations = filter_relations_by_type(graph, relations, typ)
    neighbors_attrs = [(n, graph.node[n]) for n in relations]
    for neighbor_attrs in neighbors_attrs:
        attrs = neighbor_attrs[1]
        if attr and value:
            if attr in attrs:
                val = attrs[attr]
                val = to_unicode(val)
                if val == value:
                    neighbors.append(neighbor_attrs)
        else:
            neighbors.append(neighbor_attrs)
    return neighbors

def filter_node_dict_by_value(nbunch, attr, value):
    nodes = []
    if type(nbunch) == nx.classes.multidigraph.DiGraph:
        nbunch = nbunch.nodes_iter(data=True)
    for node, attrs in nbunch:
        if attr in attrs:
            val = attrs[attr]
            if val == value:
                nodes.append((node, attrs))
    return nodes
```

### Graph

```python
castas = nx.read_gexf('castas.gexf')
```

### Characters by casta

```python
characters = node_dict_by_type(castas)[a['character']]
characters_by_casta = node_dict_by_attr(characters, a['casta'])
freq_characters_by_casta = freq_dict(characters_by_casta)
```

```python
characters_by_casta_sr = pd.Series(freq_characters_by_casta)
characters_by_casta_sr.plot(kind='bar')
```

### Spaniards by receiving action

```python
spaniards_by_receiving_action = {}
actions = map(lambda x: x[0], node_dict_by_type(castas)[a['action']])
for action in actions:
    spanish_neighbors = node_neighbor_list(
        action,
        castas,
        attr=a['casta'],
        value=a['espanol'],
        label=a['receiving'],
    )
    label = castas.node[action][a['label']]
    spaniards_by_receiving_action[label] = spanish_neighbors
freq_spaniards_by_receiving_action = freq_dict(spaniards_by_receiving_action)
```

```python
spaniards_by_receiving_action_sr = pd.Series(freq_spaniards_by_receiving_action)
spaniards_by_receiving_action_sr.plot(kind='bar')
```

### Spaniards by doing action

```python
spaniards_by_doing_action = {}
spaniards = characters_by_casta[a['espanol']]
for spaniard in spaniards:
    action_neighbors = node_neighbor_list(
        spaniard[0],
        castas,
        label=a['doing'],
    )
    for action in action_neighbors:
        label = action[1][a['label']]
        spaniards_by_doing_action.setdefault(label, [])
        spaniards_by_doing_action[label].append(spaniard)
freq_spaniards_by_doing_action = freq_dict(spaniards_by_doing_action)
actions = map(lambda x: x[1][a['label']], node_dict_by_type(castas)[a['action']])
freq_spaniards_by_doing_action = equalize_dict(actions, freq_spaniards_by_doing_action)
```

```python
spaniards_by_doing_action_sr = pd.Series(freq_spaniards_by_doing_action)
spaniards_by_doing_action_sr.plot(kind='bar')
```

### Number in title associated with Spaniards

```python
import re

num_re = r'(?P<number>\d+)'
num_comp = re.compile(num_re)
def get_number(inscription):
    result = num_comp.search(inscription)
    if result:
        number = int(result.group('number'))
    else:
        number = 'Inscription with no number'
    return number
```

```python
numbers = {}
paintings = node_dict_by_type(castas)[a['painting']]
for painting in paintings:
    spanish_neighbors = node_neighbor_list(
        painting[0],
        castas,
        attr=a['casta'],
        value=a['espanol'],
        label=a['depicts'],
        typ=a['character'],
    )
    if spanish_neighbors:
        inscription = painting[1][a['inscription']]
        if inscription:
            number = get_number(inscription)
        else:
            number = 'No inscription in the painting'
        numbers[number] = numbers.get(number, 0) + 1
```

```python
numbers_associated_with_spaniards_sr = pd.Series(numbers)
numbers_associated_with_spaniards_sr.plot(kind='bar')
```
